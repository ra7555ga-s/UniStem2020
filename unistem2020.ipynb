```python
gc = (sequence.count("G") + sequence.count("C")) / len(sequence)

print(gc)
```

```python
gc = round(gc, 4)

print(gc)
```

Or even better: 

```python
print("GC: " + str(gc*100) + "%")
```

To find where the motif is, is more complicated, but how would you solve it? We need to go through the entire genome, searching for the motif character by character in windows of the length of the motif

```python
for i in range(0, (len(sequence)-len(motif))):
    window = sequence[i:(i+len(motif))]
    if window == motif:
        print(i)
```

We can maybe split the sequence into smaller sections and look for each of them instead. Let's try it with 200 bp kmers.

```python
# This is how you subsection a string!
print(motif[1:200])
```

```python
# Print ALL THE kmers!
for i in range(len(motif)-200):
    print(str(i+1) + ".-    " + motif[i:(i+200)])
```

It would be very unlikely that we would have 200 bp exactly the same without being a match... Let's see if we can find any hits!

```python
for i in range(len(sequence)-200):
    window = sequence[i:(i+200)]
    for j in range(len(motif)-200):
        kmer = motif[j:(j+200)]
        if window == kmer:
            print("Found it!")
            break
    else:
        continue
    break
```
